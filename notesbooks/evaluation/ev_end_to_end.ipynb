```python
from evaluation.end2end_evl import evaluate_end2end
from utils.loader import GLOBAL_SCENE_GRAPH
from data.robots2 import robots_withConfig
from planner.planner_a_start import a_star_path_planner
scene_graph = GLOBAL_SCENE_GRAPH
robot_DB =robots_withConfig
robot_storage_room_id = robot_DB["robot_storage"]["room_id"]
```

```python
# the simplest trial
# simple / direct / 1 node distance / single task / wall

user_description = "The wall ID 1055 needs to be cleaned with cleaning solution."
gpt_model = "gpt-4o"
ground_truth = {
    "defect_id": 1055,
    "room_id": 45,
    "robot_id": 1,
    "path": a_star_path_planner(scene_graph, robot_storage_room_id, 45)
}
print(f"ground truth: {ground_truth}")
evaluate_end2end(user_description, scene_graph, robot_DB, gpt_model)
```

```python
# the second trial
# simple(wtih a child condition) / indirect / 3 node distance / single task / wall

user_description = "The paint of the wall with a winodw in the room ID 43 is worn out."
gpt_model = "gpt-4o"
ground_truth = {
    "defect_id": 1031,
    "room_id": 43,
    "robot_id": 3,
    "path": a_star_path_planner(scene_graph, robot_storage_room_id, 43)
}
print(f"ground truth: {ground_truth}")
evaluate_end2end(user_description, scene_graph, robot_DB, gpt_model)
```

```python
# the third trial
# complex (dimension reasoning) / indirect / 4 node distance / single task / floor

user_description = "the floor in The largest room has a crack"
gpt_model = "gpt-4o"
ground_truth = {
    "defect_id": 3026,
    "room_id": 40,
    "robot_id": 4,
    "path": a_star_path_planner(scene_graph, robot_storage_room_id, 40)
}
print(f"ground truth: {ground_truth}")
evaluate_end2end(user_description, scene_graph, robot_DB, gpt_model)
```

```python
# the fourth trial
# complex (connection reasoning) / indirect / 6 node distance / single task / door

user_description = "The door handle in room ID 13 that connects to room ID 14 is contaminated with germs."
gpt_model = "gpt-4o"
ground_truth = {
    "defect_id": 5004,
    "room_id": 13,
    "robot_id": 2,
    "path": a_star_path_planner(scene_graph, robot_storage_room_id, 13)
}
print(f"ground truth: {ground_truth}")
evaluate_end2end(user_description, scene_graph, robot_DB, gpt_model)
```

```python
# the fifth trial
# complex (mutiple tasks) / direct / 9 node distance / multiple task / walls

user_description = "all walls in room ID 32 needs to be filled and painted again."
gpt_model = "gpt-4o"
ground_truth = {
    "defect_id": [1042, 1092, 1136, 1239],
    "room_id": 32,
    "robot_id": 9,
    "path": a_star_path_planner(scene_graph, robot_storage_room_id, 32)
}
print(f"ground truth: {ground_truth}")
evaluate_end2end(user_description, scene_graph, robot_DB, gpt_model)
```
